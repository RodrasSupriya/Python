```python
import math
num=145
temp=num
s=0
while num>0:
  rem=num%10
  s=s+math.factorial(rem)
  num=num//10
if temp==s:
  print(f'{temp} it is a strong number')
else:
  print(f'{temp} it is not a strong number')
```
