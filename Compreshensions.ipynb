```python
fruits={'apple':100,'banana':200,'carrots':300}
print({i:fruits[i]*2 for i in fruits if fruits[i]>150})
```

```python
fruits={'apple':100,'banana':200,'carrots':300}
print({i:( j*3 if j>150 else j*2) for i,j in fruits.items()})
```

**Comprehension Examples**

```python
i=[i**3 for i in range(1,11)]
print(i)
```

```python
prime_list=[x for x in range(2,20)if all(x%y!=0 for y in range(2,x+1))]

print(prime_list)
```

```python
a={i:i**3 for i in range(1,6)}
print(a)
```
