```python
def factorial(n):
  if n==0 or n==1:
    return 1
  else:
    return n*factorial(n-1)
factorial(6)
```

```python
def vowels(s):
  count=0
  if s in 'aeiou':
    count+=1
  return count
print(vowels('supriya'))
```

```python
def palindrome(s):
  rev=''
  for i in s:
    rev=i+rev
  if rev==s:
    return 'palindrome'
  else:
    return 'notpalindrome'
palindrome('supriya')
```

```python
def last_three_characters(s):
  a=s[-3:]
  return a
print(last_three_characters('supriya'))
```

```python
def remove_spaces(s):
  c=[]
  for i in s:
    if i==' ':
      pass
    else:
      c.append(i)
  return ''.join(c)
print(remove_spaces('su pri ya re nuka'))
```
