```python
import json
fruits={'apple':100,'banana':200,'carrot':300}
with open ('supriya.json','w') as s:
  json.dump(fruits,s)
```

```python
fruits={'apple':100,'banana':200,'carrot':300}
with open ('supriya.json','r') as s:
  result=json.load(s)
  print(result)
```

```python
fruits={'apple':100,'banana':200,'carrot':300}
result=json.dumps(fruits)
print(result)
```

```python
final=json.loads(result)
print(final)
```
