```python
a='supriya'
for i in a:
  print(i)
```

```python
a='supriya'
i=0
while i<len(a):
  print(a[i],i)
  i=i+1
```

```python
a='supriya'
count=0
for i in a:
  if i in ('aeiou' or 'AEIOU'):
    count+=1
print(count)
```

```python
a='supriya'
i=len(a)-1
res=[]
while i>=0:
  res.append(a[i])
  i=i-1
print(''.join(res))


```

```python
a=' su priya'
b=[]
for i in a:
  if i!=' ':
    b.append(i)
print(''.join(b))

```

```python
a='aaabb'
i=0
b=[]
while i<len(a):
     if a[i] not in b:
       b.append(a[i])
       b.append(str(a.count(a[i])))
     i=i+1
print(''.join(b))
```

```python
a='SupRiYA'
b=[]
for i in a:
  if i.isupper():
    b.append(i)
print(''.join(b))
```

```python
a='s122ddddddddddd'
count=0
for i in a:
  if i.isdigit():
    count+=1
print(count)
```

**For loop using list**

```python
a=['s','u','p','r','i','y','a']
for i in a:
  print(i)
```

```python
a=['s','u','p','r','i','y','a']
i=0
while i<len(a):
  print(a[i])
  i=i+1
```

```python
a=[1,2,3]
sum=0
for i in a:
  sum+=i
print(sum)
```

```python
a=[1,122,3,6,54]
i=1
max=a[i]
while i<len(a):
  if max<a[i]:
    max=a[i]
  i=i+1
print(max)
```

```python
a=[1,122,3,6,54]
for i in a:
  if i%2==0:
    print(i)
```

```python
a=[1,122,3,6,54]
i=0
res=[]
while i<len(a):
  if a[i]%2!=0:
    res.append(a[i])
  i=i+1
print(res)
```

**For loop using dict**

```python
d={'name':'supriya','class':'btech','age':20}
count=0
for values in d.values():
  if type(values)==int:
    count+=1
print(count)
```
